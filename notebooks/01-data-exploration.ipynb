```python
import string
```

```python
def preprocess(summaries):
    for id, summary in enumerate(summaries):
        summary = summary.lower()

        chars = [char for char in summary if char not in string.punctuation]
        summaries[id] = "".join(chars)
    return summaries
```

```python
test_str = ['Hello! My, name is!! Unique.. Karki']
cleaned = preprocess(test_str)
print(cleaned[0])
```
